```python
import pandas as pd

files = {
    "2019-20": "pedri2019.xlsx",
    "2020-21": "pedri2020.xlsx",
    "2021-22": "pedri2021.xlsx",
    "2022-23": "pedri2022.xlsx",
    "2023-24": "pedri2023.xlsx",
    "2024-25": "pedri2025(ongoing).xlsx"
}

dfs = []

for season, file in files.items():
    # Read the excel file without a header
    df_raw = pd.read_excel(f"/content/{file}", header=None)

    header_row_index = None
    # Search for 'Comp' column in the first few rows to identify the actual header
    for i in range(min(5, len(df_raw))):
        # Check if 'Comp' exists in the current row's values (case-sensitive)
        if 'Comp' in df_raw.iloc[i].values:
            header_row_index = i
            break

    if header_row_index is None:
        print(f"Warning: Could not find 'Comp' in header row for {file}. Skipping this file.")
        continue

    # Set the identified row as the header
    df_temp = df_raw.iloc[header_row_index:].copy()
    df_temp.columns = df_temp.iloc[0]
    df_temp = df_temp[1:].reset_index(drop=True)

    # Ensure column names are strings (sometimes they are mixed types after read_excel)
    df_temp.columns = df_temp.columns.astype(str)

    # Check again if 'Comp' exists after header processing
    if 'Comp' not in df_temp.columns:
        print(f"Error: 'Comp' column missing after header processing for {file}. Skipping.")
        continue

    # remove repeated header rows
    df_temp = df_temp[df_temp['Comp'] != 'Comp']

    # convert minutes column properly
    df_temp['Min'] = pd.to_numeric(df_temp['Min'], errors='coerce')

    # remove invalid rows (like totals, blanks, etc.)
    df_temp = df_temp[df_temp['Min'] <= 120]

    # add season column
    df_temp['Season'] = season

    dfs.append(df_temp)

# combine all seasons into one dataframe
df = pd.concat(dfs, ignore_index=True)

# calculate average minutes per season
season_avg = df.groupby('Season')['Min'].mean()

print(season_avg)
```

```python
import pandas as pd

files = {
    "2019-20": "pedri2019.xlsx",
    "2020-21": "pedri2020.xlsx",
    "2021-22": "pedri2021.xlsx",
    "2022-23": "pedri2022.xlsx",
    "2023-24": "pedri2023.xlsx",
    "2024-25": "pedri2025(ongoing).xlsx"
}

dfs = []

for season, file in files.items():
    df_temp = pd.read_excel(f"/content/{file}", header=1)

    df_temp.columns = [str(col).strip() for col in df_temp.columns]
    df_temp = df_temp.loc[:, ~df_temp.columns.duplicated()]

    if 'Comp' in df_temp.columns:
        df_temp = df_temp[df_temp['Comp'] != 'Comp']

    min_col = None
    for col in df_temp.columns:
        if 'Min' in str(col):
            min_col = col
            break

    if min_col is None:
        print(f"Skipping {season} (no Min column found)")
        continue

    df_temp['Min'] = df_temp[min_col]
    df_temp['Min'] = df_temp['Min'].astype(str).str.extract(r'(\d+)')
    df_temp['Min'] = pd.to_numeric(df_temp['Min'], errors='coerce')

    df_temp = df_temp[df_temp['Min'] <= 120]

    df_temp['Season'] = season
    dfs.append(df_temp)

df = pd.concat(dfs, ignore_index=True)

season_avg = df.groupby('Season')['Min'].mean()
print(season_avg)
```

```python
import pandas as pd


involvement = df.groupby('Season').size()
involvement_norm = involvement / involvement.max()
avg_involvement = involvement_norm.mean()

fitness = involvement_norm.apply(
    lambda x: "High Involvement" if x > avg_involvement else "Low Involvement"
)

df['Involvement_Type'] = df['Season'].map(fitness)



team = "Osasuna"

team_high = df[
    (df['Opponent'] == team) &
    (df['Involvement_Type'] == "High Involvement")
]


if len(team_high) == 0:
    team_high = df[df['Involvement_Type'] == "High Involvement"]


pred = team_high[['Gls','Ast','Sh','TklW','Int']].mean()


prediction = pd.DataFrame(pred, columns=["Expected_per_match"]).T

prediction["Goal_Prob_%"] = prediction["Gls"] * 100
prediction["Assist_Prob_%"] = prediction["Ast"] * 100

prediction["Shots_range"] = prediction["Sh"].apply(lambda x: "0-1" if x < 1 else "1-2")
prediction["Def_Actions_range"] = (prediction["TklW"] + prediction["Int"]).apply(
    lambda x: "1-2" if x < 2 else "2-3"
)

print(prediction)
```
